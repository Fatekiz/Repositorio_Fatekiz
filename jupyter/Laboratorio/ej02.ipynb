# 2. Crear un gráfico de un vector aleatorio de 10 elementos utilizando un gráfico de líneas. (50 Pts)


#### A. Crear un array aleatorio de 10 elementos. Los valores deben ser enteros entre el intervalo -5 al 5.

```python
# importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
# creando array aleatorio del -5 al 5

aleatorio = np.random.randint(-5,6,10) # genero numeros aleatorios del -5 al 5 y que sean un total de 10 (es un vector)
aleatorio
```

#### B. Implementar un gráfico de líneas simples a partir del vector obtenido. El gráfico debe tener:

##### ● Marcadores por cada punto
#### ● Etiqueta para el gráfico
#### ● Nombre del gráfico, y nombrar los ejes X e Y

```python
grafico = plt.plot(aleatorio, label= "Vector de numeros aleatorios", marker="o") #asi creamos un grafico sencillo / como en html el label (etiqueta) / marker crea puntos en las intersec.
plt.title("Grafico de números aleatorios") # le da un titulo al grafico
plt.ylabel("Filas") # x/y label hace las etiquetas para los ejes
plt.xlabel("Columnas") 
plt.legend() # esto hace posible que se oueda ver el "label" 
plt.grid() # hacemos cuadrilla
```
