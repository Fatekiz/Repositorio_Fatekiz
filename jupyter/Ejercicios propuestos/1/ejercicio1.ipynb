#### Generar dos vectores aleatorios de valores enteros (del 1 al 10). Los vectores solamente deben tener 5 elementos y se deben sumar sus elementos para obtener un Ãºnico vector


```python
# llamando bibliotecas
import numpy as np
```

```python
vector1 = np.random.randint(1,11,5)
vector1
```

```python
vector2 = np.random.randint(1,11,5)
vector2
```

```python
# sumando los vectores
suma_vectores = vector1 + vector2
suma_vectores
```

# Utilizar la biblioteca de matplotlib para graficar el vector previamente generado 

```python
#importando lib matplotlib
import matplotlib.pyplot as plt
```

```python
plt.plot(suma_vectores, label = "vector")
plt.title("Grafico de la suma de vectores")
plt.grid(True)
plt.legend()
```
