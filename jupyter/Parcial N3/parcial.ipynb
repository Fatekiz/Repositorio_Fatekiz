# B. 1.1 Contar todos los valores nulos existentes en el dataset e imprimirlos

```python
valores_nulos = data.isnull().any().sum()
print(f"Total de valores nulos: '{valores_nulos}' ")
```

# B. 1.2 Eliminar los valores nulos existentes

```python
data_sin_null = data.dropna()
print(f"\n \n Se puede apreciar que de 2000 datos, ahora solo hay 1999 (se elimino una fila nula)  {data_sin_null.info()}") # se elimino una fila con un valor nulo
```

```python
promedio_duracion = data_sin_null['duration_ms'].mean()
print(f"El promedio de la duracion de las canciones (ms) es de: {promedio_duracion} ms")
```
